```python
data = pd.read_csv('/home/devops/Documents/Practice-Python/shakespear data - analysis/Shakespeare_data.csv')
```

```python
# in place substitution of nan in data (dataframe)
data['Player'].replace(np.nan ,'Other' , inplace = True)
```

```python
# visualising plays as network
```

```python
# first creating an empty graph with no edges and no nodes
g = nx.Graph()
```

```python
g = nx.from_pandas_dataframe(data , source = 'Play'  , target = 'Player' )
```

```python
plt.figure(figsize = (200,200))
```

```python
nx.draw_networkx(g  , arrows = True ,   with_labels = True , node_size = 58 , node_color = "g" , font_size= 100)
```

```python
plt.savefig("Graph.png", format="PNG")
```

```python
# Social Network Analysis,finding most influential play of shakespear 
# using the concept of degree centrality [play] 
# to know the counts of links held by each node [here Play].
# link https://cambridge-intelligence.com/keylines-faqs-social-network-analysis/
# We can summarize it as most influential play
```

```python
#return type dictionary
nx.degree_centrality(g)
```

```python
#return type of nx.degree_centrality is dictionary
central = pd.DataFrame(nx.degree_centrality(g) , index=[0]).T
```

```python
# assigning columns to dtaframe central
```

```python
central.columns =  ['Degree Centrality']
```

```python
central
```

```python
central['PageRank'] = pd.DataFrame(nx.pagerank(g) , index = [0]).T
```
